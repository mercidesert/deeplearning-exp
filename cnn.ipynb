```python
import numpy as np

# ==========================================
# 0. 准备数据 (和上一个例子完全一样)
# ==========================================
X = np.random.rand(11, 11, 4)     # 输入图像：11x11，4个通道
W0 = np.random.rand(5, 5, 4)      # 第0个卷积核
W1 = np.random.rand(5, 5, 4)      # 第1个卷积核

stride = 2
filter_size = 5
out_h = 4  # 输出的高度 (11-5)/2+1 = 4
out_w = 4  # 输出的宽度 (11-5)/2+1 = 4


W_weights = np.stack([W0, W1], axis=0) 
W_mat = W_weights.reshape(2, -1) 
X_col = np.zeros((100, out_h * out_w)) 
col_idx = 0 
for y in range(out_h):
    for x in range(out_w):
        y_start = y * stride
        y_end = y_start + filter_size
        x_start = x * stride
        x_end = x_start + filter_size
        patch = X[y_start:y_end, x_start:x_end, :]
        X_col[:, col_idx] = patch.flatten()
        col_idx += 1
Out_mat = np.dot(W_mat, X_col)
Out_3d = Out_mat.reshape(2, out_h, out_w)
V = Out_3d.transpose(1, 2, 0)
```
