```python
#导入库和工具环境准备
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn import datasets
from sklearn.model_selection import train_test_split

%matplotlib inline
plt.rcParams['figure.figsize']=(6.0,4.0)
plt.rcParams['image.interpolation']='nearest'
plt.rcParams['image.cmap']='gray'

%load_ext autoreload
%autoreload 2
```

```python
digits=datasets.load_digits()
X=digits.data
Y=digits.target
images=digits.images

print('图片总数：',X.shape[0])
print('每张图片像素个数',X.shape[1])

plt.imshow(images[0])
plt.title(f"标准答案：{Y[0]}")
plt.axis('off')
plt.show()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)

print('训练集数量：',X_train.shape[0])
print('测试集数量：',X_test.shape[0])
```

```python
from knn import knn
classifier=knn()
classifier.train(X_train,Y_train)
test_pred=classifier.predict(X_test,k=3)

num_correct=np.sum(test_pred==Y_test)
total=X_test.shape[0]
accuracy=float(num_correct)/total
print(f'准确率为{accuracy}')
```
